```python
import matplotlib.pyplot as plt
import numpy as np

# Dati per P1
labels = ['Min.', 'Max.', 'Avg.', 'Totale']
values_P1 = [1.7, 34, 16, 40.9]
values_P1_costs = [493, 968, 559, 279500]
values_P1_resources = [99.65, 99.65, 3.64, 1.57]

# Dati per P1'
values_P1_prime = [1.2, 32.8, 14.2, 40.5]
values_P1_prime_costs = [493, 793, 532, 265750]
values_P1_prime_resources = [99.65, 99.65, 3.68, 1.84]

# Larghezza degli istogrammi
bar_width = 0.35

# Creazione dei grafici
fig, axs = plt.subplots(1, 3, figsize=(15, 5))  # Modificato il numero di righe e colonne

# Grafico per la distribuzione dei tempi di ciclo
bar_positions = np.arange(len(labels))
axs[0].bar(bar_positions - bar_width / 2, values_P1, bar_width, color='lightblue', label='P1')
axs[0].bar(bar_positions + bar_width / 2, values_P1_prime, bar_width, color='lightcoral', label="P1'", alpha=0.7)
axs[0].set_title('Distribuzione dei Tempi di Ciclo')
axs[0].set_xticks(bar_positions)
axs[0].set_xticklabels(labels)
axs[0].legend()

# Grafico per la distribuzione dei costi
bar_positions = np.arange(len(labels) - 1)  # Rimuovi l'ultimo elemento
axs[1].bar(bar_positions - bar_width / 2, values_P1_costs[:-1], bar_width, color='lightblue', label='P1')
axs[1].bar(bar_positions + bar_width / 2, values_P1_prime_costs[:-1], bar_width, color='lightcoral', label="P1'", alpha=0.7)
axs[1].set_title('Distribuzione dei Costi')
axs[1].set_xticks(bar_positions)
axs[1].set_xticklabels(labels[:-1])  # Rimuovi l'ultimo elemento dalle etichette
axs[1].legend()
axs[1].set_ylabel('Costi (€)')  # Aggiungi l'unità di misura all'asse y
axs[1].tick_params(axis='x', which='both', bottom=False)  # Nascondi le etichette sull'asse x

# Grafico per l'utilizzo delle risorse
resource_labels = ['A1', 'S1', 'M1', 'M2']
resource_positions = np.arange(len(resource_labels))
axs[2].bar(resource_positions - bar_width / 2, values_P1_resources, bar_width, color='lightblue', label='P1')
axs[2].bar(resource_positions + bar_width / 2, values_P1_prime_resources, bar_width, color='lightcoral', label="P1'", alpha=0.7)
axs[2].set_title('Utilizzo delle Risorse')
axs[2].set_xticks(resource_positions)
axs[2].set_xticklabels(resource_labels)
axs[2].legend()

# Aggiungi bordi neri agli istogrammi
for ax in axs:
    for container in ax.containers:
        for bar in container:
            bar.set_edgecolor('black')
            bar.set_linewidth(1.5)  # Larghezza del bordo

# Regola gli spazi tra i subplots
plt.tight_layout()

# Mostra i grafici
plt.show()
```
